```python
def detect_outliers_iqr(df):
    # calculate the IQR for each column in the DataFrame
    Q1 = df.quantile(0.25)
    Q3 = df.quantile(0.75)
    IQR = Q3 - Q1
    
    # define a multiplier for the IQR
    multiplier = 1.5
    
    # use numpy.logical_or function to identify the indices of the outliers in the DataFrame
    lower_bound = Q1 - multiplier * IQR
    upper_bound = Q3 + multiplier * IQR
    outlier_indices = np.logical_or(df < lower_bound, df > upper_bound)
    
    # create a new DataFrame that only contains the outliers
    outliers_df = df[outlier_indices]
    
    return outliers_df
```

```python
outliers_df = detect_outliers_iqr(df)
```

```python
outliers_df.count()
```
